# Análise de Métodos de Sorteamento

A seguir seguem uma anális de gráficos para três diferentes tipos de sorteamento, sendo eles: SelectionSort, InsertionSort, BubbleSort. Antes da análise geral dos dados coletados, há um gráfico feito para cada PC que rodou o código de sorteamento. No final, há uma análise sobre o resultado obtido.

```python
#Import de bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Leitura de dados
dados = pd.read_csv("Dados.csv")
df = pd.DataFrame(dados)
df.head(10)
```

```python
#Dados gerais
tipos = ["selectionSort","insertionSort","bubbleSort"]
quantidade = [40000,80000,160000,320000,640000,1280000,2560000]

```

```python
#Dados DESKTOP-LSTUBER

media_lstuber_selection = []
media_lstuber_insertion = []
media_lstuber_bubble = []
dados_lstuber = dados.loc[dados["pc"] == "DESKTOP-LSTUBER", :]

for tipo in tipos:
    for qntd in quantidade:
        dados_selection = dados_lstuber.loc[dados_lstuber["tipo"] == tipo, :]
        dados_selection = dados_selection.loc[dados_selection["tamanho"] == qntd, :]
        if (tipo=="selectionSort"):
            media_lstuber_selection.append(dados_selection["tempo"].mean())
        elif tipo == "insertionSort":
            media_lstuber_insertion.append(dados_selection["tempo"].mean())
        else:
            media_lstuber_bubble.append(dados_selection["tempo"].mean())

plt.plot(quantidade,media_lstuber_selection, label="SelectionSort")
plt.plot(quantidade,media_lstuber_insertion, label="InsertionSort")
plt.plot(quantidade,media_lstuber_bubble, label="BubbleSort")
plt.xlabel("Tempo de execução (ms)")
plt.ylabel("Quantidade de pontos")
plt.title("Média LSTUBER de execução dos métodos de sorteamento")
plt.show()
```
